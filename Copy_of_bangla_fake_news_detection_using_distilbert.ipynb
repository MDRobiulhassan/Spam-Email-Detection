# Create Distilbert Model With GRU Layer

```python
# def create_model():
#     input_shape = Input(shape = (max_len,), dtype='int32')
#     masks= Input(shape = (max_len,), dtype='int32')
#     distil_bert_layer = distil_bert_model(input_ids=input_shape, attention_mask=masks)[0]
#     X = (tf.keras.layers.GRU(128, return_sequences=True))(distil_bert_layer)
#     X = tf.keras.layers.GlobalMaxPool1D()(X)
#     X=(tf.keras.layers.Dense(64,activation="tanh"))(X)
#     X=(tf.keras.layers.Dense(32))(X)


#     X = tf.keras.layers.Dense(16)(X)

#     X = tf.keras.layers.Dense(8)(X)

#     X = tf.keras.layers.Dense(4)(X)
#     X = tf.keras.layers.Dropout(0.5)(X)
#     X = tf.keras.layers.Dense(num_classes, activation='softmax')(X)
#     model = tf.keras.Model(inputs=[input_shape,masks], outputs=X)






#     return model

import tensorflow as tf
from tensorflow.keras.layers import Input, GRU, Dense, Dropout
from tensorflow.keras.models import Model  # Import Model
from transformers import TFDistilBertModel

class DistilBertLayer(tf.keras.layers.Layer):
    def __init__(self, **kwargs):
        super(DistilBertLayer, self).__init__(**kwargs)
        self.distil_bert = TFDistilBertModel.from_pretrained('distilbert-base-uncased')
        self.distil_bert.trainable = True  # Set to True to fine-tune DistilBERT

    def call(self, inputs, attention_mask):
        outputs = self.distil_bert(input_ids=inputs, attention_mask=attention_mask)[0]
        return outputs

def create_model():
    input_shape = Input(shape=(max_len,), dtype='int32')
    masks = Input(shape=(max_len,), dtype='int32')

    distil_bert_layer_output = DistilBertLayer()(input_shape, masks)

    X = GRU(128, return_sequences=True)(distil_bert_layer_output)
    X = tf.keras.layers.GlobalMaxPool1D()(X)
    X = Dense(64, activation="tanh")(X)  # Using Dense
    X = Dense(32)(X)  # Using Dense
    X = Dense(16)(X)
    X = Dense(8)(X)
    X = Dense(4)(X)
    X = Dropout(0.5)(X)
    X = Dense(num_classes, activation='softmax')(X)

    model = Model(inputs=[input_shape, masks], outputs=X)
    return model


```
